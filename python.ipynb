Q3 Compare and contrast mutable and immutable objects in Python with examples
Ans Mutable Objects - Objects can be changed in place without changing their
     identity (memory address). Common mutable types include lists dictionaries, and sets.
     Example:-
              my_list = [1, 2, 3]
           my_list.append(4)
           print(my_list)

     Immutable Objects - Immutable objects cannot be changed after creation. If you try to modify them, a new object is created. Common immutable types include integers, floats, strings, and tuples.
     Example:-
                name = "Alice"
             name = name + " Smith"
             print(name)  # Output: Alice Smith
             
Q4  Discuss the different types of operators in Python and provide examples of
     how they are used.
Ans Arithmetic Operators- Used for basic mathematical operations.
     Examples:
      a + b    # Addition  
      a - b    # Subtraction  
      a * b    # Multiplication  
      a / b    # Division  
      a // b   # Floor division  
      a % b    # Modulus (remainder)  
      a ** b   # Exponentiation  

     Comparison (Relational) Operators-Used to compare two values and return a Boolean result (True or False).
     Examples:

      a == b   # Equal to  
      a != b   # Not equal to  
      a > b    # Greater than  
      a < b    # Less than  
      a >= b   # Greater than or equal to  
      a <= b   # Less than or equal to  

      Assignment Operators-Used to assign values to variables and perform compound operations.
      Examples:
          x = 10    # Assign  
          x += 5    # Add and assign (x = x + 5)  
          x -= 3    # Subtract and assign  
          x *= 2    # Multiply and assign  
          x /= 4    # Divide and assign  

      Logical Operators-Used to combine multiple conditions.
       Examples:
        a > 5 and a < 10    # Logical AND  
        a < 5 or a > 10     # Logical OR  
        not(a > 5)          # Logical NOT  

       Bitwise Operators-Perform operations on binary values.
       Examples:
        a & b    # Bitwise AND  
        a | b    # Bitwise OR  
        a ^ b    # Bitwise XOR  
        ~a       # Bitwise NOT  
        a << 1   # Left shift  
        a >> 1   # Right shift  

Q5 Explain the concept of type casting in Python with examples
Ans Type casting in Python refers to the process of converting one data type into another. This is useful when you need to perform operations that require variables to be in a specific type, such as converting a string to an integer before doing arithmetic. Python provides built-in functions for type casting like int(), float(), str(), and bool().

Types of Type Casting:
Implicit Type Casting

Done automatically by Python when no data loss occurs.

Example:

python
Copy
Edit
x = 10       # int
y = 2.5      # float
result = x + y
print(result)        # Output: 12.5
print(type(result))  # Output: <class 'float'>
Here, x (int) is automatically converted to float for the operation.

Explicit Type Casting

Done manually using type conversion functions.

Examples:

python
Copy
Edit
# String to Integer
a = "100"
b = int(a)
print(b + 50)  # Output: 150

# Float to Integer
x = 3.7
y = int(x)
print(y)       # Output: 3 (decimal part is removed)

# Integer to String
num = 10
text = str(num)
print(text + " apples")  # Output: 10 apples
